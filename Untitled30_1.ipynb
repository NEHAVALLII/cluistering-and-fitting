Load the Data

```python
# Import necessary libraries
import pandas as pd

# Load the dataset
file_path ='/content/Car details v3.csv'  # Replace with your file path
data = pd.read_csv(file_path)

# Preview the data
print(data.head())

# Summary of dataset
print(data.info())
print(data.describe())
```

 Cleaning and Preparing the Dataset

```python
import numpy as np

# Drop irrelevant columns
data = data.drop(columns=["name", "torque"])

# Clean mileage, engine, and max_power columns (extract numeric values)
data['mileage'] = data['mileage'].str.extract('(\d+\.\d+|\d+)').astype(float)
data['engine'] = data['engine'].str.extract('(\d+)').astype(float)
data['max_power'] = data['max_power'].str.extract('(\d+\.\d+|\d+)').astype(float)

# Handle missing values
data['mileage'].fillna(data['mileage'].mean(), inplace=True)
data['engine'].fillna(data['engine'].mean(), inplace=True)
data['max_power'].fillna(data['max_power'].mean(), inplace=True)
data['seats'].fillna(data['seats'].mode()[0], inplace=True)

# Encode categorical variables
categorical_columns = ['fuel', 'seller_type', 'transmission', 'owner']
data = pd.get_dummies(data, columns=categorical_columns, drop_first=True)

# Normalize numerical columns
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
numerical_columns = ['year', 'selling_price', 'km_driven', 'mileage', 'engine', 'max_power', 'seats']
data[numerical_columns] = scaler.fit_transform(data[numerical_columns])

# Check cleaned dataset
print(data.head())
```

```python
# Handle missing values explicitly (no inplace=True)
data['mileage'] = data['mileage'].fillna(data['mileage'].mean())
data['engine'] = data['engine'].fillna(data['engine'].mean())
data['max_power'] = data['max_power'].fillna(data['max_power'].mean())
data['seats'] = data['seats'].fillna(data['seats'].mode()[0])

# Check the cleaned dataset again
print(data.head())
```

Correlation Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Compute the correlation matrix
correlation_matrix = data.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

Distribution of Fuel Types

```python
import matplotlib.pyplot as plt

# Calculate the fuel type counts
fuel_counts = data[['fuel_Diesel', 'fuel_LPG', 'fuel_Petrol']].sum().rename({
    'fuel_Diesel': 'Diesel',
    'fuel_LPG': 'LPG',
    'fuel_Petrol': 'Petrol'
})

# Pie Chart for Fuel Types
fuel_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['blue', 'orange', 'green'])
plt.title("Fuel Type Distribution")
plt.ylabel("")  # Hide the y-label
plt.show()
```

Box Plot

```python
import seaborn as sns

# Box plot for selling price distribution
plt.figure(figsize=(8, 6))
sns.boxplot(x=data['selling_price'], color='lightblue')
plt.title('Selling Price Distribution')
plt.xlabel('Selling Price (in INR)')
plt.show()
```

 Scatter Plot

```python
import matplotlib.pyplot as plt

# Scatter plot of selling price vs km driven
plt.figure(figsize=(8, 6))
plt.scatter(data['km_driven'], data['selling_price'], alpha=0.5, color='purple')
plt.title('Selling Price vs Kilometer Driven')
plt.xlabel('Kilometers Driven')
plt.ylabel('Selling Price (in INR)')
plt.show()
```

Elbow Plot

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Elbow Method
wcss = []  # Within-cluster sum of squares
for i in range(1, 11):  # Try 1 to 10 clusters
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(data)  # Assuming `data` is the normalized DataFrame
    wcss.append(kmeans.inertia_)

# Plot Elbow Method
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

Silhouette Score

```python
from sklearn.metrics import silhouette_score

# Test silhouette score for 2 to 10 clusters
silhouette_scores = []
for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    cluster_labels = kmeans.fit_predict(data)
    silhouette_scores.append(silhouette_score(data, cluster_labels))

# Plot Silhouette Scores
plt.plot(range(2, 11), silhouette_scores, marker='o', linestyle='--', color='orange')
plt.title('Silhouette Score for Optimal Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.show()
```

Applying K-Means Clustering

```python
# Fit K-Means with the optimal number of clusters
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)
data['Cluster'] = kmeans.fit_predict(data)

# Cluster Centers
centers = kmeans.cluster_centers_
print("Cluster Centers:\n", centers)
```

Visualization of Clusters

```python
plt.figure(figsize=(8, 6))
for cluster in range(3):  # Assuming k=3
    cluster_data = data[data['Cluster'] == cluster]
    plt.scatter(cluster_data['selling_price'], cluster_data['mileage'], label=f'Cluster {cluster}')

# Plot cluster centers
plt.scatter(centers[:, 0], centers[:, 1], s=200, c='red', marker='X', label='Centroids')
plt.title('Clusters of Cars')
plt.xlabel('Selling Price')
plt.ylabel('Mileage')
plt.legend()
plt.show()
```
